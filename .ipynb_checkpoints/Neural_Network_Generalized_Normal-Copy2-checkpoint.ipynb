```python
#change to 1 label to learn beta
import tensorflow as tf
import numpy as np
import Normal_Distribution_Data as data
import matplotlib.pyplot as plt
```

```python
#load input data for gaussian
training = data.create_training_set(1000, 100) #outputs an array of x_vals, mean values, and sigma values respectively
x_train = training[0] 
mu_train = training[1]  
alpha_train = training[2]
beta_train = training[3] 
y_train = np.vstack((mu_train, alpha_train, beta_train)) #compile training mean and standard deviation into single data structure
y_train = y_train.T #first column as mean, second column as sd

testing = data.create_training_set(300, 100)
x_test = testing[0] 
mu_test = testing[1] 
alpha_test = testing[2] 
beta_test = testing[3] 
y_test = np.vstack((mu_test, alpha_test, beta_test)) #compile test mean and standard deviation into single data structure 
y_test = y_test.T

```

```python
#set layers
model = tf.keras.models.Sequential() #feed forward
#input layer: 100 different x values to represent the gaussian function
model.add(tf.keras.layers.Input(100))
#hidden layers (5):
model.add(tf.keras.layers.Dense(200, activation = tf.nn.relu))   
model.add(tf.keras.layers.Dense(200, activation = tf.nn.relu))
model.add(tf.keras.layers.Dense(200, activation = tf.nn.relu))
#model.add(tf.keras.layers.Dense(200, activation = tf.nn.relu))
#model.add(tf.keras.layers.Dense(125, activation = tf.nn.relu))
#output layer:
model.add(tf.keras.layers.Dense(3,)) #3 labels: mu, alpha, and beta
```

```python
#optimize
model.compile(optimizer = 'adam', 
              loss = 'mean_squared_error',
              metrics = [tf.keras.metrics.RootMeanSquaredError()])
model.fit(x_train, y_train, epochs = 575) #90 epochs minimizes the difference between test and training and testing cost
```

```python
#evaluate the test set and find test loss and accuracy
results = model.evaluate(x = x_test, y = y_test)
val_loss = results[0] #loss between model and actual
val_acc = results[1] #accuracy of the model based on actual
print(val_loss, val_acc)
```

```python
#saves the model data
model.save('generalized_normal.model')
new_model = tf.keras.models.load_model('generalized_normal.model') 
```

```python
#makes predictions for all elements in x_test array  
predictions = new_model.predict(x_test) #predicts all means and standard deviations of the x_test dataset 
```

```python
predictions = predictions.T
y_test = y_test.T
plt.scatter(predictions[0],y_test[0]) #mean
plt.title('Predicted Mu vs. Actual Mu')
plt.xlabel('Predicted Mu')
plt.ylabel('Actual Mu')
plt.show() 
```

```python
plt.scatter(predictions[1],y_test[1]) #SD
plt.title('Predicted Alpha vs. Actual Alpha')
plt.xlabel('Predicted Alpha')
plt.ylabel('Actual Alpha')
plt.show()
```

```python
plt.scatter(predictions[2],y_test[2]) #SD
plt.title('Predicted Beta vs. Actual Beta')
plt.xlabel('Predicted Beta')
plt.ylabel('Actual Beta')
plt.show()
```
